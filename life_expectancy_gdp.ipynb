```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('all_data.csv')
# Display the column names
print(data.columns)
```

```python
# Display first few rows
print(data.head(5))
```

```python
# rename Life expectancy at birth (years)
data.rename(columns= {'Life expectancy at birth (years)' : 'Life'}, inplace=True)
print(data.head(5))
```

```python
# Descriptive statistics
print(data.describe())
```

```python
# Find unique countries
unique_countries = data['Country'].unique()
print(f"Unique countries: {unique_countries}")
print(f"Number of unique countries: {len(unique_countries)}")
```

```python
# Scatter plot
sns.scatterplot(x='GDP', y='Life',hue='Country', data=data)
plt.title('GDP vs Life Expectancy by Country')
plt.show()
```

Insight:The scatter plot illustrates the relationship between GDP and life expectancy across the selected countries. 
It appears that as GDP increases, life expectancy tends to increase as well, indicating a positive correlation.

```python
# Line plot for GDP over time
sns.lineplot(x='Year', y='GDP', hue='Country', data=data)
plt.title('GDP over Time by Country')
plt.show()
```

Insight: The line plot shows the trends in GDP over time for each country. 
Notably, the United States displays the highest GDP throughout the years, while Zimbabwe shows a more erratic trend.


```python
# Line plot for Life Expectancy over time
sns.lineplot(x='Year', y='Life', hue='Country', data=data)
plt.title('Life Expectancy over Time by Country')
plt.show()
```

Insight: The line plot for life expectancy indicates that life expectancy has generally increased over time for most countries. 
However, the rate of increase varies among the countries, with Chile and Germany showing a steady rise.

```python
# Select relevant columns
gdp_data = data[['Country', 'GDP']]
gdp_data = gdp_data.groupby('Country').mean().reset_index()

# Sort by GDP from min to max
gdp_data_sorted = gdp_data.sort_values(by='GDP', ascending=True)

# Create a bar plot with color
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='GDP', y='Country', data=gdp_data_sorted, palette='viridis',hue ='Country')

# Add title and labels
plt.title('GDP per Country (Sorted from Minimum to Maximum)')
plt.xlabel('GDP (in USD or appropriate unit)')
plt.ylabel('Country')
plt.show()
```

Insight: This bar plot represents the average GDP of each country, sorted from minimum to maximum. 
It is evident that the United States has the highest GDP, while Zimbabwe has the lowest.

```python
# Select relevant columns
life_expectancy_data = data[['Country', 'Life']]

# Create a box plot to visualize the distribution of life expectancy
plt.figure(figsize=(12, 6))
sns.boxplot(x='Life', data=data, palette='viridis',hue='Country',legend='auto')

# Add title and labels
plt.title('Distribution of Life Expectancy Across Countries')
plt.xlabel('Life Expectancy (in years)')
plt.show()
```

Insight: The box plot illustrates the distribution of life expectancy across the selected countries. 
The interquartile range (IQR) highlights the middle 50% of values, indicating that most countries have a life expectancy between 74 and 79 years.

In conclusion, this analysis shows a clear positive correlation between GDP and life expectancy. 
Countries with higher GDP tend to have higher life expectancy. The visualizations provided insights into trends over time,as well as the distribution of life expectancy across the selected countries. Overall, economic factors appear to play a significant role in health outcomes.
